````python
k_A=0
n_A=0
k_G=0
n_G=0
k_C=0
n_C=0
for i in range (df_penguins_full.shape[0]):
  if(df_penguins['species'].loc[df_penguins.index[i]]=="Adelie"):
    k_A=k_A+df_penguins['body_mass_g'].loc[df_penguins.index[i]]
    n_A=n_A+1
  if(df_penguins['species'].loc[df_penguins.index[i]]=="Gentoo"):
    k_G=k_G+df_penguins['body_mass_g'].loc[df_penguins.index[i]]
    n_G=n_G+1
  if(df_penguins['species'].loc[df_penguins.index[i]]=="Chinstrap"):
    k_C=k_C+df_penguins['body_mass_g'].loc[df_penguins.index[i]]
    n_C=n_C+1
avg_A=k_A/n_A
avg_G=k_G/n_G
avg_C=k_C/n_C
p=abs(avg_A-avg_G)
q=abs(avg_C-avg_G)
r=abs(avg_A-avg_C)
MIN=min(p,q,r)
if(MIN==p):
  print("Adelie and Gentoo")
if(MIN==q):
  print("Chinstrap and Gentoo")
if(MIN==r):
  print("Adelie and Chinstrap")
````

**40.** We have a hypothesis that as the day goes on, the tips get higher.  We'll need to wrangle the data a bit before testing this, however.  First, now that we have a datetime column, pickup_time, create a subset of it to create a new DataFrame, taxis_one_day. This new DataFrame should have values between '2019-03-23 00:06:00' (inclusive) and '2019-03-24 00:00:00' (exlusive).

````python
taxis_one_day= taxis[(taxis['pickup'] >= '2019-03-23 00:06:00') & (taxis['pickup'] <  '2019-03-24 00:00:00')]

print(taxis_one_day)
````

**41.** We now have a range from morning until midnight, but we to take the mean of the numeric columns, grouped at one hour intervals.  Save the result as df_means, and display it.

````python
taxis_one_day[''] = taxis_one_day['pickup_time'].dt.hour
DF_means = taxis_one_day.groupby('').mean()
DF_means
````

**44.** Create a new column in taxis_means, ```tip_in_percent```.  The source columns for this should be "fare" and "tip"

````python
tip_in_percent_list = []
time_interval_list=[]
for i in range(0,24):
  tip_percent = (DF_means.iloc[i, 3]/DF_means.iloc[i, 2])*100
  tip_in_percent_list.append(tip_percent)
  time_interval_list.append(i)
tip_in_percent_arr = np.array(tip_in_percent_list)
time_interval_arr = np.array(time_interval_list)
DF_means['tip_in_percent'] = tip_in_percent_arr
DF_means['time_interval'] = time_interval_arr
DF_means
````
